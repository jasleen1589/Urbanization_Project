```python
#dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
```

```python
#setting file path
file = Path("../Resources/population_density.csv")
population = pd.read_csv(file)
population.head()
```

```python
#renaming columns for this table
renamed_population = population.rename(columns={"name":"County", "landArea": "Area/sq mi"})
renamed_population.head()
```

```python
#reading in second file
file2 = Path("../Resources/co-est2019-annres-06.csv")
population2 = pd.read_csv(file2)
population2.head()
```

```python
#removing the overall california data
population3 = population2[population2["Area"] != "California"]
population3.head()
```

```python
#renaming the columns for the second table
renamed_population3 = population3.rename(columns={"Area":"County", "2019": "2019 Population Estimate"})
renamed_population3.head()
```

```python
#merging the dataframes
merge_df = pd.merge(renamed_population, renamed_population3, on="County")
merge_df.head()
```

```python
#selecting only the columns i want to look at
cleaned_df = merge_df[["County", "2019 Population Estimate", "Area/sq mi"]]
cleaned_df.head()
```

```python
#removing the commas in the 2019 population estimate column
cleaned_df.replace(',', '', regex=True, inplace=True)
cleaned_df.head()
```

```python
#checking the datatypes
cleaned_df.dtypes
```

```python
#changing the 2019 population estimate column values to a float so we can perform calculations on it
cleaned_df = cleaned_df.astype({"2019 Population Estimate": float}, errors='raise')
```

```python
#checking the datatypes again to make sure it worked
cleaned_df.dtypes
```

```python
#creating a new column to find density per/sq mi, you do this by dividing the population by
#the the land area (in this case is square miles)
density = (cleaned_df["2019 Population Estimate"]/cleaned_df["Area/sq mi"])
cleaned_df["Density/sq mi"] = density

cleaned_df.head()
```

```python
#rounding to two decimal places
cleaned_df[["Density/sq mi"]]=cleaned_df[["Density/sq mi"]].round(2)
cleaned_df.head()
```

```python
#sorting the data to ascending so it complies with plotly
sorted_population = cleaned_df.sort_values("Density/sq mi")
sorted_population.head()
```

```python
sorted_population = sorted_population.reset_index(drop = True)
sorted_population
```

```python
sorted_population.to_csv("../Resources/population_density_v1.csv", index = False, header = True)
```
