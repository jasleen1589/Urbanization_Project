```python
#dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
import seaborn as sns
```

```python
#calling in the file path
file = Path("../Resources/homelessness_vs_density.csv")
df = pd.read_csv(file)
df.head()
```

```python
#testing a simple scatterplot using seaborn
sns.scatterplot(x="Density/sq mi", y="Experiencing Homelessness %", data=df)
plt.show()
```

```python
#creating a scatter plot with continuous sizes and hues
sns.set_theme(style="whitegrid")

# Load the homeless_rate dataset
homeless_rate = df

cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
g = sns.relplot(
    data=homeless_rate,
    x="Density/sq mi", y="Experiencing Homelessness %",
    hue="Experiencing Homelessness %", size="Experiencing Homelessness %",
    palette=cmap, sizes=(100, 200),
)
g.set(xscale="linear", yscale="linear")
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)

#plt.savefig("seaborn_plot.png")
```
